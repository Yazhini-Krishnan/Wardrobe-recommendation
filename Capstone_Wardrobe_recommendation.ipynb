```python
import tensorflow
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from keras.layers import GlobalMaxPooling2D


```

```python
model=ResNet50(weights='imagenet',include_top=False, input_shape=(224,224,3))
model.trainable=False
```

```python
model=tensorflow.keras.Sequential([
    model,
    GlobalMaxPooling2D()
])
model.summary()
```

```python
import cv2
import numpy as np
from numpy.linalg import norm
```

```python
img=cv2.imread("/content/drive/MyDrive/Dataset/33088.jpg")
img=cv2.resize(img,(224,224))
```

```python
img=np.array(img)
img.shape
```

```python
expand_img=np.expand_dims(img, axis=0)
expand_img.shape
```

```python
pre_img=preprocess_input(expand_img)
#pre_img
#pre_img.shape
```

```python
result=model.predict(pre_img).flatten()
normalised=result/norm(result)
#result.shape
normalised.shape
```

```python
def extract_features(img_path, model):
  img=cv2.imread(img_path)
  img=cv2.resize(img,(224,224))
  img=np.array(img)
  expand_img=np.expand_dims(img, axis=0)
  pre_img=preprocess_input(expand_img)
  result=model.predict(pre_img).flatten()
  normalised=result/norm(result)
  return normalised



```

```python
extract_features("/content/drive/MyDrive/Dataset/33086.jpg", model)
```

```python
import os
from tqdm import tqdm
Dataset="/content/drive/MyDrive/Dataset"
filename=[]
feature_list=[]

for file in os.listdir(Dataset):
  filename.append(os.path.join(Dataset, file))

#filename  
```

```python
for file in tqdm(filename):
  feature_list.append(extract_features(file,model))
```

```python
import pickle

pickle.dump(feature_list,open('featurevector.pk1','wb'))
pickle.dump(filename,open('filename.pk1','wb'))
```
